```python

#1.kutuphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#2.veri onisleme
#2.1.veri yukleme
veriler = pd.read_csv('veriler.csv')
#pd.read_csv("veriler.csv")
#test
print(veriler)

x= veriler.iloc[ : , 2:4 ].values # bağımsız değişkenler
y= veriler.iloc [: , 4:].values # Bağımlı değişkenler


#verilerin egitim ve test icin bolunmesi
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)

#verilerin olceklenmesi
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)


#-------------------------------
# Sınıflandırmaya Başlayalım
from sklearn.linear_model import LogisticRegression
Logreg=LogisticRegression(random_state=0)
Logreg.fit(X_train,y_train)
y_pred=Logreg.predict(X_test)
print(y_pred)
print(y_test)


#Confusion Matrisi -  Karışıklık Matrisi
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)


#KNN  Algoritması
from sklearn.neighbors import KNeighborsClassifier
# Burada n_neighbors değerini yüksek tutmamın sebebi klasik düşünce olan verisetinin ne kadar çok veriyle eğitilirse sonuçları o kadar iyi tahmin eder yaklaşımını kırmakm için bunu belirtmek istedim verileri azaltarak doğruluk değerimizin komşu değerlere bağlı olduğunu veri sayısına bağlı olmadığını göstermek istedim

knnAlgoritm=KNeighborsClassifier(n_neighbors=7,metric="minkowski")
knnAlgoritm.fit(X_train,y_train)
y_pred=knnAlgoritm.predict(X_test)
cm= confusion_matrix(y_test,y_pred)
print(cm)

knnAlgoritm=KNeighborsClassifier(n_neighbors=1,metric="minkowski")
knnAlgoritm.fit(X_train,y_train)
y_pred=knnAlgoritm.predict(X_test)

cm= confusion_matrix(y_test,y_pred)
print(" KNN Algoritm ")
print(cm)


#-------------------------
# Naive Byes  Gaussian Naive Bayes

from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
y_pred=gnb.predict(X_test)
cm= confusion_matrix(y_test,y_pred)
print("Gaussion Naive Bayes ")
print(cm)


from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(X_train,y_train)
y_pred=bnb.predict(X_test)

cm= confusion_matrix(y_test,y_pred)
print("Bernoulli Naive Bayes")
print(cm)
```
