```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

veriler=pd.read_csv('Musteriler.csv')
print(veriler)

X = veriler.iloc[: , 3:].values
print(X)

from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4,init="k-means++",random_state=182)
kmeans.fit(X)

print(kmeans.cluster_centers_)
sonuclar=[]

for i in range(1,10):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=213)
    kmeans.fit(X)
    sonuclar.append(kmeans.inertia_)

plt.plot(range(1,10),sonuclar)
plt.show()

#--------------------------------------
#Farklı cluster değerleriyle deneme
kmeans2=KMeans(n_clusters=5,init="k-means++",random_state=100)
kmeans2.fit(X)
print(kmeans2.cluster_centers_)
sonuclar2=[]
for j in range(1,20):
    kmeans2=KMeans(n_clusters=j,init="k-means++",random_state=90)
    kmeans2.fit(X)
    sonuclar2.append(kmeans2.inertia_)

plt.plot(range(1,20),sonuclar2)
plt.show()
# K-MEANS 3 (Random) ile
sonuclar3=[]
kmeans3=KMeans(n_clusters=2,init="random")
kmeans3.fit(X)
print(kmeans3.cluster_centers_)

for k in range(1,30):
    kmeans3=KMeans(n_clusters=2,init="random",random_state=90)
    kmeans3.fit(X)
    sonuclar3.append(kmeans3.inertia_)

plt.plot(range(1,30),sonuclar3)
plt.show()


#K-Means kendi girdiğimiz merkezi değerlerle
kmeans4=KMeans(n_clusters=3,init="k-means++")
kmeans4.fit(X)
sonuclar4=[]
def my_init(X, n_clusters, random_state=40):
    # Örnek olarak ilk n veriyi merkez olarak al
    return X[:n_clusters]

kmeans4 = KMeans(n_clusters=2, init=my_init, random_state=90)
kmeans4.fit(X)
sonuclar4.append(kmeans4.inertia_)

print("sonuclar4:", sonuclar4)
plt.scatter([0],sonuclar4)
plt.plot([0],sonuclar4)
plt.show()



#K-Means farklı değerler(Random ile)
kmeans5 = KMeans(n_clusters=7,init="random")
kmeans5.fit(X)
sonuclar5=[]

for i in range(1,50):
    kmeans5=KMeans(n_clusters=7,init="random")
    kmeans5.fit(X)
    sonuclar5.append(kmeans5.inertia_)

plt.plot(range(1,50),sonuclar5)
plt.show()


#------------------------------
#K-Means++ ile Fraklı değerlerin yorumu
kmeans6=KMeans(n_clusters=7,init="k-means++")
kmeans6.fit(X)
print(kmeans6.labels_)
sonuclar6=[]
for r in range(1,100):
    kmeans6=KMeans(n_clusters=7,init="k-means++")
    kmeans6.fit(X)
    sonuclar6.append(kmeans6.inertia_)

plt.plot(range(1,100),sonuclar6)
plt.show()


#------------------
kmeans7=KMeans(n_clusters=3,init="random")
kmeans7.fit(X)
print(kmeans7.labels_)
sonuclar7=[]
for p in range(1,200):
    kmeans7=KMeans(n_clusters=4,init="random")
    kmeans7.fit(X)
    sonuclar7.append(kmeans7.inertia_)

plt.plot(range(1,200),sonuclar7)
plt.show()


#--------------------------
kmeans8=KMeans(n_clusters=4,init="k-means++")
kmeans8.fit(X)
print(kmeans8.labels_)
sonuclar8=[]
for l in range(1,200):
           means8=KMeans(n_clusters=4,init="k-means++")
           means8.fit(X)
           sonuclar8.append(means8.inertia_)

plt.plot(range(1,200),sonuclar8)
plt.show()



```
